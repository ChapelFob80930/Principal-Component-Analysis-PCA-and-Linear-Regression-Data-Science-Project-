#2.3 Principal Component Analysis (PCA)

```python
plt.boxplot(dataset.drop('DateTime', axis=1))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
ds1 = dataset.drop(['DateTime','GenDiffFlows',
       'DiffFlows'], axis=1)
scaler = StandardScaler()
std_data = scaler.fit_transform(ds1)
print(ds1.head)
print(np.var(ds1['Temperature']))
print(np.var(ds1['Humidity']))
print(np.var(ds1['WindSpeed']))
print(np.var(ds1['PCZone1']))
print(np.var(ds1['PCZone2']))
print(np.var(ds1['PCZone3']))
```

```python
plt.boxplot(std_data)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit(std_data)
results = pca.transform(std_data)
```

```python
print(pca.explained_variance_)
per_var= np.round(pca.explained_variance_ratio_*100,decimals=1)
print(per_var)
print(np.cumsum(per_var))
```

```python
loadings = pca.components_.T
print("loadings")
print(ds1.columns)
print(loadings)
```

```python
import matplotlib.pyplot as plt
labels =["PC1","PC2","PC3","PC4","PC5","PC6"]
plt.plot([1,2,3,4,5,6,],per_var)
plt.ylabel("Percentage of variance explained (PVE)")
plt.xlabel("Principal component")
plt.title("Scree plot")
```

```python
import matplotlib.pyplot as plt
labels =["PC1","PC2","PC3","PC4","PC5","PC6"]
plt.plot([1,2,3,4,5,6,],np.cumsum(per_var))
plt.ylabel("Percentage of variance explained (PVE)")
plt.xlabel("Principal component")
plt.title("Scree plot")
```

```python
#correlation circle
loadings = pca.components_.T
fig,axis = plt.subplots(figsize=(5,5))
axis.set_xlim(-1,1)
axis.set_ylim(-1,1)
plt.plot([-1,1],[0,0],color="silver",linestyle="-",linewidth=1)
plt.plot([0,0],[-1,1],color="silver",linestyle="-",linewidth=1)
for j in range(0,6):
    plt.arrow(0, 0, loadings[j,0], loadings[j,1],head_width=0.02,width=0.001,color="red")
    plt.annotate(ds1.columns[j],(loadings[j,0], loadings[j,1]))
circle = plt.Circle((0,0),1,color='blue',fill=False)
axis.add_artist(circle)
```

# **Linear Regression**

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# target = dataset['PCZone1']
# print(target.head())
```

```python
predictors = dataset.drop(['DateTime','GenDiffFlows',
       'DiffFlows'], axis=1)
print(predictors.head())
```

```python
predictors.corr()["PCZone1"]
```

```python
lm = sm.OLS.from_formula('PCZone1 ~ PCZone2', dataset)
tetuan_regsimple = lm.fit()
```

```python
print(tetuan_regsimple.params)
```

```python
# Assuming 'predictors.PCZone2' contains the data for PCZone2
x_min = predictors.PCZone2.min()
x_max = predictors.PCZone2.max()

# Regression coefficients
intercept = 6631.589167
slope = 1.246530

# Calculate corresponding y values
y_min = intercept + slope * x_min
y_max = intercept + slope * x_max

print(x_min)
print(x_max)
print(y_min)
print(y_max)
# Updated plot with better endpoints
plt.scatter(predictors.PCZone2, predictors.PCZone1)
plt.plot([0, x_max], [6631.589167, y_max], '-', color='red')  # Regression line
plt.xlabel("PCZone2")
plt.ylabel("PCZone1")
plt.title("Power Consumption in Zone 1 (PCZone1) vs. Power Consumption in Zone 2 (PCZone2)")
plt.show()
```

```python
tetuan_regsimple.conf_int(alpha=0.05)
```

# **Multiple Linear Regression**

```python
from itertools import combinations as com
```

```python
target = dataset['PCZone1']
```

```python
predictors= dataset.drop(['DateTime','GenDiffFlows',
       'DiffFlows','PCZone1'], axis=1)
```

```python
print(predictors.head())
```

```python
for n in range(1,6)
  print("For n = ",n)
  combination = com(predictors.columns, n)
  out = [i for i in combination]
  print("X1 = ",out[0][0])
  print("X2 = ",out[0][1])
  lm = sm.OLS.from_formula('PCZone1 ~ out[0][0]+out[0][1]', predictors)
  tetuan_regMultiple = lm.fit()
  print(tetuan_regMultiple.summary())
```

```python
# Identify predictors and target variable
target = 'PCZone1'
  # Drop the target column to retain predictors

# Initialize variables to store the best model
best_model = None
best_adj_r2 = -float('inf')  # Start with negative infinity
best_combination = None

# Iterate over the number of features (1 to 5)
for n in range(1, 6):
    print(f"\nFor n = {n}")  # Number of predictors used

    # Generate all combinations of n predictors
    combinations = itertools.combinations(predictors.columns, n)

    # Loop through each combination
    for subset in combinations:
        print("Predictors used:", subset)

        # Create the formula for the OLS model
        formula = f"{target} ~ {' + '.join(subset)}"

        # Fit the model using OLS
        lm = sm.OLS.from_formula(formula, dataset)
        model = lm.fit()

        # Print the model summary for each combination


        # Check if this model has the highest adjusted R
        if model.rsquared_adj > best_adj_r2:
            best_model = model
            best_adj_r2 = model.rsquared_adj
            best_combination = subset

# Output the best model
print("\nBest Model Selected:")
print("Predictors:", best_combination)
print("Adjusted R:", best_adj_r2)
print(best_model.summary())

# Coefficients
print("\nCoefficients of the best model:")
print(best_model.params)

# Interpretation of Coefficient of Determination (R)
print("\nR of the best model:", best_model.rsquared)
print(f"Interpretation: {best_model.rsquared:.2%} of the variability in {target} is explained by the selected predictors.")

# Perform zero slope hypothesis test for all coefficients except 0 (Intercept)
print("\nZero Slope Hypothesis Tests:")
for predictor in best_combination:
    p_value = best_model.pvalues[predictor]
    print(f"P-value for {predictor}: {p_value}")
    if p_value < 0.05:
        print(f"{predictor} is statistically significant.")
    else:
        print(f"{predictor} is not statistically significant.")

# Make predictions with the best model for given conditions
new_data = pd.DataFrame({
    'Temperature': [26],   # Replace with actual predictor names and values
    'Humidity': [65],
    'WindSpeed': [4.2],
    'PCZone2': [18840],
    'PCZone3': [25700]
})
prediction = best_model.predict(new_data)
print("\nPredicted PCZone1:", prediction[0])
```
